```python
import nltk
nltk.download("all") #selects the entire set of book resources
from nltk import word_tokenize
from nltk.sentiment import SentimentIntensityAnalyzer

#Importing webscraping modules and libraries
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlencode
import matplotlib.pyplot as plt
from urllib.request import urlopen as uReq
```

```python
url =  'https://www.amazon.com/Amazon-Halo-Fitness-And-Health-Band/product-reviews/B07QJ6YL4T/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber='
```

```python
url_list = []
for i in range(1,500):
    url_list.append(url+str(i))
```

```python
url_list[:5]
```

```python
reviews = []
rating = []    
title = []


for i,url in enumerate(url_list):
    print("\npage ", i+1,":") #Keeps track of page numbers
    params = {'api_key': "cdb7b58bbbdf56ac54b5957bc7b19afd", 'url':url }
    
    response = requests.get('http://api.scraperapi.com/',params=urlencode(params))
    soup = BeautifulSoup(response.text, 'html.parser')
    data_string = ""
    
    #Scraping review content contained in each page corresponding to the titles above
       
    #Keeps track of page numbers
    for item in soup.find_all("span", {"data-hook": "review-body"}):
          data_string = data_string + item.get_text()
          reviews.append(data_string.strip())
          data_string= ""
            
    for item in soup.find_all("a", {"data-hook": "review-title"}):
          data_string = data_string + item.get_text()
          title.append(data_string.strip())
          data_string= ""
        
    rate_string = ""
    for item in soup.find_all("i", {"data-hook": "review-star-rating"}):
        rate_string = item.get_text()
        rating.append(rate_string[:1])
    
    print(len(title))
```

```python
print(rating)
```

```python
import pandas as pd

final = pd.DataFrame([reviews,rating],index=['Review','Rating'])
final = final.T
```

```python
final.to_csv('Smart_Watch_Review.csv',header=True)
```
